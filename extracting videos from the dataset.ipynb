```python
import json
import os
import shutil

# Load the JSON file
with open('WLASL_v0.3.json', 'r') as file:
    data = json.load(file)

# Define the path to the videos folder and the parent folder for gloss folders
videos_folder = './videos_dataset/videos'
parent_folder = 'dataset'

# Create the parent folder if it doesn't exist
if not os.path.exists(parent_folder):
    os.makedirs(parent_folder)

# Iterate through the first 100 entries in the JSON data
for entry in data[:100]:
    gloss = entry['gloss']
    instances = entry['instances']
    gloss_folder = os.path.join(parent_folder, gloss)

    # Create a folder for the gloss if it doesn't exist
    if not os.path.exists(gloss_folder):
        os.makedirs(gloss_folder)

    # Iterate through each instance to collect the videos
    for instance in instances:
        video_id = instance['video_id']
        source_url = instance['url']
        video_filename = f"{video_id}.mp4"
        source_filename = os.path.join(videos_folder, video_filename)

        # Check if the video file exists in the videos folder
        if os.path.exists(source_filename):
            # Move the video file to the gloss folder
            destination_filename = os.path.join(gloss_folder, video_filename)
            shutil.move(source_filename, destination_filename)
            print(f"Moved video {video_id} to {gloss_folder}")
        else:
            print(f"Video {video_id} not found for gloss {gloss}")

print("Videos collection completed.")

```
